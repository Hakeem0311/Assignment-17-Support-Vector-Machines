## Prepare a classification model using SVM for Forest Fires


```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, cross_val_score
```

## Data collection and description


```python
forest=pd.read_csv('forestfires (2).csv')
forest
```

```python
X= forest.iloc[:,0:30]
Y= forest.iloc[:,30]
X
```

```python
Y
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X= X.apply(LabelEncoder().fit_transform)
X
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=0.3, random_state=40)
```

## SVM (Support Vector Machine )


1.Linear

```python
#LINEAR SUPPORT VECTOR MACHINE
model_linear = SVC(kernel= "linear")
model_linear.fit(x_train,y_train)
preds_linear = model_linear.predict(x_test)
np.mean(preds_linear==y_test)
```

2.poly

```python
#POLY SUPPORT VECTOR MACHINE
model_poly = SVC(kernel="poly")
model_poly.fit(x_train,y_train)
preds_poly = model_poly.predict(x_test)
np.mean(preds_poly==y_test)
```

3.RBF


```python
#RBF SUPPORT VECTOR MACHINE
model_rbf = SVC(kernel="rbf")
model_rbf.fit(x_train,y_train)
preds_rbf=model_rbf.predict(x_test)
np.mean(preds_rbf==y_test)
```

```python
from sklearn.metrics import classification_report
model=SVC()
model.fit(x_train,y_train)
pred=model.predict(x_test)
print(classification_report(pred,y_test))
```

## Grid Search CV RBF


```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10],'C':[15,14,13,12] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_
```
